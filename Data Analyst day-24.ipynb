# Implorting Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from datetime import datetime
```

```python
df=pd.read_csv('Netflix.csv')
```

```python
df
```

# First 5 Data

```python
df.head()
```

```python
sns.set(rc={'figure.figsize':(10,5)})
```

```python
df['Date']=pd.to_datetime(df['Date'])
df=df.set_index('Date')
df.head()
```

```python
sns.lineplot(x=df.index,y=df['Volume'],label="volume",color="green")
plt.title("Volume of Stock verses time")
```

```python
df.plot(y=['High','Close','Open'],title="Netflix Stock Price",color=["Red","blue","white"])
```

```python
fig,(ax1,ax2,ax3) = plt.subplots(3,figsize=(15,10))
df.groupby(df.index.day).mean().plot(y='Volume',ax=ax1,xlabel='Day',color="violet")
df.groupby(df.index.month).mean().plot(y='Volume',ax=ax2,xlabel='Month',color='Orange')
df.groupby(df.index.year).mean().plot(y='Volume',ax=ax3,xlabel='Year',color='Brown')
```

# Dates With Highest Stock price

```python
a=df.sort_values(by='High',ascending=False).head(5)
a['High']
```

# Data With Lowest Stock Price

```python
b=df.sort_values(by='Low',ascending=True).head(5)
b['Low']
```

```python
Fig,axes=plt.subplots(nrows=1,ncols=2,sharex=True,figsize=(10,5))
fig.suptitle("High and Low values Stock per period of time",fontsize=18)
sns.lineplot(ax=axes[0],y=df['High'],x=df.index,color='green')
sns.lineplot(ax=axes[1],y=df['Low'],x=df.index,color='Red')
```
